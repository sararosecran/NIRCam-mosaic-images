```python
#function reads in file
def read_file(filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                print(line.rstrip())
    except IOError as e:
        print(e)
```

```python
#puts all fits into np arrays
hdu_list_x1 = fits.open('x1.example.fits')
hdu_list_x2 = fits.open('x2.example.fits')
hdu_list_xout = fits.open('xout1.example.fits')
hdu_list_y1 = fits.open('y1.example.fits')
hdu_list_y2 = fits.open('y2.example.fits')
hdu_list_yout = fits.open('yout1.example.fits')
tbdat_x1=hdu_list_x1[0].data
tbdat_x1 = tbdat_x1.flatten()
tbdat_x2=hdu_list_x2[0].data
tbdat_x2 = tbdat_x2.flatten()
tbdat_xout=hdu_list_xout[0].data
tbdat_xout = tbdat_xout.flatten()
tbdat_y1=hdu_list_y1[0].data
tbdat_y1 = tbdat_y1.flatten()
tbdat_y2=hdu_list_y2[0].data
tbdat_y2 = tbdat_y2.flatten()
tbdat_yout=hdu_list_yout[0].data
tbdat_yout = tbdat_yout.flatten()
```

```python
#plot coordinates
fig = plt.figure(figsize=(8,8))
fig.suptitle('Pixel Locations')
plt.plot(tbdat_x1, tbdat_y1, marker = "o", color='red') #plots exposure 1 
#plt.plot(tbdat_x2, tbdat_y2, marker = "s", color='blue') #plots exposure 2
#plt.plot(tbdat_xout, tbdat_yout, marker = "v", color = 'green') #plots output pixel locations
plt.xlim((-1,65))
plt.ylim((-1,35))
plt.show()
```
